## 我的步驟
1. Forward Path Gain (找到向前的路 - 前饋路徑增益 P1, P2, ....)
2. Individual Loop Gain (找到每個loop - 環路增益 L1, L2, ....)
3. 總體 delta = 1 - (L1+L2+...+Ln) + Σ(兩兩獨立的環乘積) + Σ (三三獨立的環乘積) + ...
4. delta K (各路徑delta)
   - delta 1 = 1 - Σ (與第一條路徑 P1 獨立的環) + Σ (與第一條路徑 P1 獨立的兩兩環乘積) + ...
   - delta 2 = 1 - Σ (與第一條路徑 P2 獨立的環) + Σ (與第一條路徑 P2 獨立的兩兩環乘積) + ...
   - ...
5. do the formula
   - Transfer function  = Σ (P1路徑增益*Δ1路徑1的delta, P2增益*Δ2路徑2delta, P3增益*Δ3路徑3delta, .... ) / 總體 Δ

## 印度人的步驟
1. Forward Path Gain (找到向前的路 - 前饋路徑增益)
2. Individual Loop Gain (找到每個loop - 環路增益)
3. gain product of two non-touching loops 
4. delta and delta K
5. do the formula
