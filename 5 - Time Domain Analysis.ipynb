Stability 與 Performance 是回授控制系統中很重要的問題。

古典控制理論中，研究performance (系統性能)方法其中兩種
- time response
- frequency response

此章便談 time response (時間響應)

觀察控制系統，隨時間增加，其 time response 變化的情形，並對此加以分析，就稱為**時域分析(Time Domain Analysis)**
(分析 time response)

# Time Response 時間響應 #

控制系統輸入沒有一定型式，故規劃一些 【標準測試訊號】 來探討系統性能

| 標準測試訊號 | 物理意義 | 表示式  | 拉式轉換 |
|---|---|---|---|
| 單位步階函數 (Unit Step Function) | 位置控制 | Us(t) | $\frac{1}S$ |
| 單位斜坡函數 (Unit Ramp Function) | 速度控制 | $t\cdot U_s(t)$ | $\frac{1}{S^2}$ |
| 單位拋物線函數 (Unit Parabolic Function) | 加速度控制 | ${\frac{t^2}2}\cdot U_s(t)$ | $\frac{1}{S^3}$ |

- 步階:測試系統 對 輸入瞬間變動 的 響應敏捷性 
- 斜坡:測試系統 對 隨時間做線性變化 的 輸入響應 (定速度運動)
- 拋物線: 固定的加速度

## Time Response 分兩部分 ##

C (t) =  Ct (t) + Css (t)

### Transient Response 暫態響應 ###
Ct (t) : 系統達到穩態之前的響應<BR>
$\lim_{t\to \infty} C_t (t) = 0$

### Steady-State Response 穩態響應 ###
Css (t) : 系統達到穩態後的響應 <BR>
$\lim_{t\to \infty} C(t) = C_{ss} (t)$
